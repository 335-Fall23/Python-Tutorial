## Dataframe from Dictionary

Dictionary is a built-in Python Data structure.
We can create Pandas dataframe from dictionary

```python
#Dictionary is a built-in Python Data structure

dict = {"country": ["Brazil", "Vietnam", "India", "China", "South Africa"],
       "capital": ["Brasilia", "Hanoi", "New Dehli", "Beijing", "Pretoria"] }

print("Dictionary :\n\n", dict)

#Creating Pandas dataframe from dictionary

df = pd.DataFrame(dict)

print("\nDataFrame :\n", df)
```

```python
#print column 'country'
country=df["country"]
print(country)
```

```python
#print columns 'country' and 'capital'

country_capital= df[['country', 'capital']]
print(country_capital)
```

## Dataframe from List

List is a built-in Python Data structure.
We can create Pandas dataframe from List

```python
L =["Brazil", "Vietnam", "India", "China", "South Africa"]

print("List :\n\n", L)

#Creating Pandas series from dictionary

countries = pd.Series(L)

print("\nSeries :\n", countries)
```

```python
L2 =[["Brazil", "Brasilia"],["Vietnam","Hanoi"],["India","New Dehli"],["China","Beijing"],["South Africa", "Pretoria"]]

print("List :\n\n", L2)

#Creating Pandas series from dictionary

countries_capital = pd.Series(L2)

print("\nSeries :\n", countries_capital)
```

# Indexing Data Frames

```python
df = pd.read_csv('data/advertising.csv') 

# Print out first 4 observations
print(df[0:4])

# Print out fifth and sixth observation
print(df[4:6])
```

```python
#Location based Indexing

#print fourth column

sales=df.iloc[:,3]

print(sales)
```

## DataFrame Columns to Numpy Arrays

```python
#convert sales column to numpy array

sales_arr= sales.to_numpy()

print(sales_arr)

print("\nDimensions of sales array", sales_arr.ndim)
print('\nShape of sales array:', sales_arr.shape)
```

```python
#convert one dimensional sales array to 2-D array with 1 column and 200 rows

sales_arr= sales_arr.reshape(-1,1)

print(sales_arr)

print("\nDimensions of sales array", sales_arr.ndim)
print('\nShape of sales array:', sales_arr.shape)
```
