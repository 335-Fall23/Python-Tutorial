```python
#Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
#Read csv File and Load to a dataframe

df = pd.read_csv('data/advertising.csv')
df
```

```python
#create x,y series

x_series = df["TV"]
y_series = df["Sales"]

#draw scatter plot

plt.scatter(x_series, y_series)
plt.xlabel("TV Advertising")
plt.ylabel("Sales")

```

```python
#linear regression

# x and y are 1-D arrays. We need to convert them into 2-D arrays to fit the regression model.
# we first convert series into numpy arrays and use reshape function.

x = x_series.to_numpy().reshape(-1,1)
y = y_series.to_numpy().reshape(-1,1)

#train model
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

#plot scatterplot
plt.scatter(x,y)

#plot regression_line
plt.plot(x,y_pred, color='red', linewidth=2)
```

```python
#To retrieve the slope:
m=model.coef_
print(m)

#For retrieving the intercept:
c=model.intercept_
print(c)
```

```python
predicted_400=model.predict([[400]])
print(predicted_400)

predicted_350=model.predict([[350]])
print(predicted_350)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred, color='red', linewidth=2)
plt.scatter(400, predicted_400)
plt.scatter(350, predicted_350)
```
